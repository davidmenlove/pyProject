```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Define Simulation Parameters

n_policies = 100
n_years = 10000
severity = np.random.lognormal(mean=9,sigma=1,size=n_policies)

# Create claim probabilities

claim_probs = np.random.uniform(
    low = 0.01,
    high = 0.10,
    size = n_policies
)
```

```python
# Simulate Claims

random_draws = np.random.rand(n_years,n_policies)
```

```python
# Determine whether claims occur

claims = random_draws < claim_probs
```

```python
# Convert claim occurences into losses

losses = claims * severity
```

```python
# Aggregate annual losses

annual_losses = losses.sum(axis=1)
```

```python
# Calculate VaR

var_95 = np.percentile(annual_losses,95)
```

```python
# Calculate CVaR

tail_losses = annual_losses[
    annual_losses >= var_95]
cvar_95 = tail_losses.mean()
```

```python
# Plot the distribution

mean_loss = annual_losses.mean()

plt.hist(annual_losses, bins =30)

plt.axvline(
    mean_loss,
    linestyle = '--',
    label = f'Mean = {mean_loss:,.0f}',
    color = 'black'
)

plt.axvline(
    var_95,
    linestyle = '--',
    label = f'95% Var = {var_95:,.0f}',
    color = 'red'
)

plt.axvline(
    cvar_95,
    linestyle = '--',
    label = f'95% CVar = {cvar_95:,.0f}',
    color = 'purple'
)

plt.axvspan(var_95,max(annual_losses),color='red',alpha=.2)
plt.xlabel('Annual Loss')
plt.ylabel('Frequency')
plt.title('Simulated Annual Loss Distribution')
plt.legend()

plt.show()
```
